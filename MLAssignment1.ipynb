```python
# Python ≥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

import os
datapath = os.path.join("datasets", "housing", "")

# To plot pretty figures directly within Jupyter
%matplotlib inline
import matplotlib as mpl
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Code example
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model

# Load the data
housing = pd.read_csv(datapath + "housing.csv", thousands=',')

# Prepare the data
X = np.c_[housing["median_income"]]
y = np.c_[housing["median_house_value"]]

# Visualize the data
housing.plot(kind='scatter', x="median_income", y='median_house_value')

plt.show()
```

```python
# Select a linear model
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X, y)
t0, t1 = model.intercept_[0], model.coef_[0][0]
t0, t1
```

```python
# Visualize the linear regression
housing.plot(kind='scatter', x="median_income", y='median_house_value')

#Plot the linear regression line on trhe graph
plt.plot(X, t0 + t1*X, "b")

plt.show()
```

```python
#Predict the house value for income of 30
predict_for_income=[[30]]
print(model.predict(predict_for_income)) # outputs [[ 5.96242338]]
```
